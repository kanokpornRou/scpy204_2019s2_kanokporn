```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from astroquery.sdss import SDSS
query = "select s.z,s.zErr,petroMag_g,petroMagErr_g FROM photoObjAll JOIN specObjAll s ON s.bestObjID = objID WHERE zWarning = 0 and s.z between 0.3 and 0.4 and petroMag_g>-5000 and petroMagErr_g<1 and s.zErr<=0.0002 "
a = SDSS.query_sql(query)
print(a)
```

```python
z=a["z"]
zErr=a["zErr"]
petroMag_g=a["petroMag_g"]
petroMagErr_g=a["petroMagErr_g"]
```

```python
fig=plt.figure(dpi=500)
plt.errorbar(z,petroMag_g,yerr=petroMagErr_g,xerr=zErr,fmt='o',markersize=2.0)
plt.title("grahp of petroMag_g vs z")
plt.xlim(0.3,0.4)
plt.xlabel("z")
plt.ylabel("petroMag_g")
```
