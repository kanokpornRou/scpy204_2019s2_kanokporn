```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('COVID-19-geographic-disbtribution-worldwide-2020-04-12.csv')
df = pd.read_csv('COVID-19-geographic-disbtribution-worldwide-2020-04-12.csv')
df
```

```python
states = df[df['countriesAndTerritories'] == 'Thailand']
states
```

```python
sumcase = states['cases'].sum()
per = (sumcase/69428524)*100
die = states['deaths'].sum()
perdie = (die/sumcase)*100
data2 = pd.Series([sumcase,per,perdie],index=['รวมผู้ติดเชื้อในไทย','เปอร์เซ็นผู้ติดเชื้อต่อประชากร','เปอร์เซ็นผู้เสียชีวิตต่อผู้ติดเชื้อ'])
data2
```

```python
data3 = pd.read_csv('COVID-19-geographic-disbtribution-worldwide-2020-04-12.csv',index_col=['geoId'])
data3
```

```python
thd=0
cnd=0
vnd=0
add=0
zwd=0
lad=0
twd=0
gmd=0
bzd=0
jpd=0
```

```python
thc=0
cnc=0
vnc=0
adc=0
zwc=0
lac=0
twc=0
gmc=0
bzc=0
jpc=0
```

```python
ath=[]
acn=[]
avn=[]
aad=[]
azw=[]
ala=[]
atw=[]
agm=[]
abz=[]
ajp=[]
```

```python
for j in range(10332) :
    i=10331-j
    countries=data3.iloc[i]['countriesAndTerritories']
    deaths=data3.iloc[i]['deaths']
    cases=data3.iloc[i]['cases']
    if countries == 'Thailand' :
        thd=thd+deaths
        thc=thc+cases
        ath.append(thc)
    if countries == 'Vietnam' :
        vnd=vnd+deaths
        vnc=vnc+cases
        avn.append(adc)
    if countries == 'Andorra' :
        add=add+deaths
        adc=adc+cases
        aad.append(adc)
    if countries == 'Zimbabwe' :
        zwd=zwd+deaths
        zwc=zwc+cases
        azw.append(zwc)
    if countries == 'Laos' :
        lad=lad+deaths
        lac=lac+cases
        ala.append(lac)
    if countries == 'Taiwan' :
        twd=twd+deaths
        twc=twc+cases
        atw.append(twc)
    if countries == 'Germany' :
        gmd=gmd+deaths
        gmc=gmc+cases
        agm.append(gmc)
    if countries == 'Brazil' :
        bzd=bzd+deaths
        bzc=bzc+cases
        abz.append(bzc)
    if countries == 'Japan' :
        jpd=jpd+deaths
        jpc=jpc+cases
        ajp.append(jpc)
    if countries == 'Canada' :
        cnd=cnd+deaths
        cnc=cnc+cases
        acn.append(cnc)
```

```python
fig=plt.figure()
plt.plot(np.arange(len(ath))+1,ath,label='Thailand')
plt.plot(np.arange(len(avn))+1,avn,label='Vietnam')
plt.plot(np.arange(len(aad))+1,aad,label='Andorra')
plt.plot(np.arange(len(azw))+1,azw,label='Zimbabwe')
plt.plot(np.arange(len(ala))+1,ala,label='Laos')
plt.plot(np.arange(len(atw))+1,atw,label='Taiwan')
plt.plot(np.arange(len(agm))+1,agm,label='Germany')
plt.plot(np.arange(len(abz))+1,abz,label='Brazil')
plt.plot(np.arange(len(ajp))+1,ajp,label='Japan')
plt.plot(np.arange(len(acn))+1,acn,label='Canada')
plt.yscale('log')
plt.legend(bbox_to_anchor=(1.02,0.7),loc='upper left')
plt.xlim(0)
plt.title("Relation graph between the number of days and number of cases in 10 countries")
plt.xlabel("Number of days")
plt.ylabel("Number of cases")
```

```python
fig=plt.figure()
plt.scatter(thd * 100/thc,thc * 100/69428524,s=694,alpha=0.3,c='tab:orange',label='Thailand')
plt.scatter(vnd * 100/vnc,vnc * 100/95540395,s=955,alpha=0.3,c='tab:pink',label='Vietnam')
plt.scatter(add * 100/adc,adc * 100/77006,s=8,alpha=0.3,c='tab:blue',label='Andorra')
plt.scatter(zwd * 100/zwc,zwc * 100/14439018,s=364,alpha=0.3,c='tab:green',label='Zimbabwe')
plt.scatter(lad * 100/lac,lac * 100/7061507,s=71,alpha=0.3,c='tab:red',label='Laos')
plt.scatter(twd * 100/twc,twc * 100/23780452,s=237,alpha=0.3,c='tab:purple',label='Taiwan')
plt.scatter(gmd * 100/gmc,gmc * 100/82927922,s=830,alpha=0.3,c='tab:brown',label='Germany')
plt.scatter(bzd * 100/bzc,bzc * 100/209469333,s=2095,alpha=0.3,c='tab:cyan',label='Brazil')
plt.scatter(jpd * 100/jpc,jpc * 100/126529100,s=1265,alpha=0.3,c='tab:olive',label='Japan')
plt.scatter(cnd * 100/cnc,cnc * 100/37058856,s=367,alpha=0.3,c='tab:gray',label='Canada')
plt.legend(bbox_to_anchor=(1.02,0.7),loc='upper left')
plt.title("Relation graph between %case/population and %death/cases in 10 countries")
plt.xlabel("%case/population(%)")
plt.ylabel("%death/case(%)")
plt.ylim(-0.1,0.8)
```
